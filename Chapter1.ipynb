```python
import numpy as np
import matplotlib.pyplot as plt

#データの作成
x = np.arange(0,6,0.1) #0 から 6まで０．１刻みで生成
y = np.sin(x)

#グラフの描画
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# データの差k末井
x = np.arange(0,6,0.1) # 0 から 6まで0.1刻みで生成
y1 = np.sin(x)
y2 = np.cos(x)

#グラフの描画
plt.plot(x,y1, label="sin")
plt.plot(x, y2,linestyle="--",label="cos") #破線で描画
plt.xlabel("x")
plt.title('sin & cos')
plt.legend()
plt.show()
```
